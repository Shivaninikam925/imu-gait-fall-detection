```python
import os
import arff
from collections import Counter

# Path to your ARFF files root
root_path = r"C:\Users\Shivani\OneDrive\Documents\B.Tech\IMU-Gait-Fall-Detection\imu-gait-fall-detection\data\raw\WISDM_ar_v1.1\arff_files"

activity_counts = Counter()

# Walk through all subfolders
for root, dirs, files in os.walk(root_path):
    for file in files:
        if file.endswith(".arff"):
            file_path = os.path.join(root, file)

            with open(file_path, 'r') as f:
                data = arff.load(f)

                # Skip empty files
                if len(data['data']) == 0:
                    continue

                for row in data['data']:
                    activity = row[-1]  # Activity label is the last column
                    activity_counts[activity] += 1

# Show class distribution
print("Activity Counts:")
for activity, count in activity_counts.items():
    print(f"{activity}: {count}")

# Simple balance check
print("\nIs dataset balanced?")

max_count = max(activity_counts.values())
min_count = min(activity_counts.values())

if max_count - min_count < 0.1 * max_count:
    print("✔ Yes, roughly balanced.")
else:
    print("✘ No, dataset is imbalanced.")
```
